```python
# 1 M -=> 1024 kb

def size_process(item):
    if str(item)[-1] == "M":
        res = float(str(item).replace("M",""))
        result = res * 1024
        return result
    elif str(item)[-1] == "k":
        res = float(str(item).replace("k",""))
        return res
    else:
        return str(np.NaN)
        
```

```python
size_process("19M")
```

```python
data_copy["Size"] = data_copy["Size"].apply(size_process)
```
