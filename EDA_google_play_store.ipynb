```python
# 1 M -=> 1024 kb

def size_process(item):
    if str(item)[-1] == "M":
        res = float(str(item).replace("M",""))
        result = res * 1024
        return result
    elif str(item)[-1] == "k":
        res = float(str(item).replace("k",""))
        return res
    else:
        return str(np.NaN)
        
```

```python
size_process("19M")
```

```python
data_copy["Size"] = data_copy["Size"].apply(size_process)
```

```python
def android_ver_process(item):
    if pd.isna(item):
        return np.nan
    else:
        item = str(item).split("-")[-1].strip()
        return '.'.join(item.split('.'))[:2]
        

data_copy['Android Ver'] = data_copy['Android Ver'].apply(android_ver_process)
data_copy['Android Ver'].unique()
```

```python
fig = plt.figure(figsize=(12,5))
sns.countplot(x = data_copy['Content Rating'], palette='colorblind')
plt.xlabel("Content Rating")
plt.show()
```

Insight from the above graph:
Most apps are in the “Everyone” category (around 8000). This shows that the majority of apps are made for all age groups.

The “Teen” category is much smaller (around 1000 apps). This means fewer apps are specifically targeted at teenagers.

Categories like “Everyone 10+” and “Mature 17+” have very limited apps compared to “Everyone”.

The “Adults only 18+” category has almost no apps. This indicates strict policies or low demand for adult-only apps.

“Unrated” apps are also very few.
